### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total por tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f'Ingreso total Tienda 1: ${ingreso_tienda1:,.2f}')
print(f'Ingreso total Tienda 2: ${ingreso_tienda2:,.2f}')
print(f'Ingreso total Tienda 3: ${ingreso_tienda3:,.2f}')
print(f'Ingreso total Tienda 4: ${ingreso_tienda4:,.2f}')
```

# 2. Ventas por categoría

```python
# Ventas por categoría en cada tienda
ventas_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("Ventas por categoría - Tienda 1:\n", ventas_categoria_tienda1)
print("\nVentas por categoría - Tienda 2:\n", ventas_categoria_tienda2)
print("\nVentas por categoría - Tienda 3:\n", ventas_categoria_tienda3)
print("\nVentas por categoría - Tienda 4:\n", ventas_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de la tienda
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f'Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}')
print(f'Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}')
print(f'Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}')
print(f'Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}')
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Tienda 1
ventas_productos_tienda1 = tienda['Producto'].value_counts()
mas_vendido_tienda1 = ventas_productos_tienda1.idxmax()
menos_vendido_tienda1 = ventas_productos_tienda1.idxmin()
print(f"Tienda 1 - Producto más vendido: {mas_vendido_tienda1} ({ventas_productos_tienda1.max()} ventas)")
print(f"Tienda 1 - Producto menos vendido: {menos_vendido_tienda1} ({ventas_productos_tienda1.min()} ventas)")
ventas_productos_tienda1.plot(kind='bar', title='Ventas por producto - Tienda 1')
plt.xlabel('Producto')
plt.ylabel('Cantidad de ventas')
plt.show()

# Tienda 2
ventas_productos_tienda2 = tienda2['Producto'].value_counts()
mas_vendido_tienda2 = ventas_productos_tienda2.idxmax()
menos_vendido_tienda2 = ventas_productos_tienda2.idxmin()
print(f"Tienda 2 - Producto más vendido: {mas_vendido_tienda2} ({ventas_productos_tienda2.max()} ventas)")
print(f"Tienda 2 - Producto menos vendido: {menos_vendido_tienda2} ({ventas_productos_tienda2.min()} ventas)")
ventas_productos_tienda2.plot(kind='bar', title='Ventas por producto - Tienda 2')
plt.xlabel('Producto')
plt.ylabel('Cantidad de ventas')
plt.show()

# Tienda 3
ventas_productos_tienda3 = tienda3['Producto'].value_counts()
mas_vendido_tienda3 = ventas_productos_tienda3.idxmax()
menos_vendido_tienda3 = ventas_productos_tienda3.idxmin()
print(f"Tienda 3 - Producto más vendido: {mas_vendido_tienda3} ({ventas_productos_tienda3.max()} ventas)")
print(f"Tienda 3 - Producto menos vendido: {menos_vendido_tienda3} ({ventas_productos_tienda3.min()} ventas)")
ventas_productos_tienda3.plot(kind='bar', title='Ventas por producto - Tienda 3')
plt.xlabel('Producto')
plt.ylabel('Cantidad de ventas')
plt.show()

# Tienda 4
ventas_productos_tienda4 = tienda4['Producto'].value_counts()
mas_vendido_tienda4 = ventas_productos_tienda4.idxmax()
menos_vendido_tienda4 = ventas_productos_tienda4.idxmin()
print(f"Tienda 4 - Producto más vendido: {mas_vendido_tienda4} ({ventas_productos_tienda4.max()} ventas)")
print(f"Tienda 4 - Producto menos vendido: {menos_vendido_tienda4} ({ventas_productos_tienda4.min()} ventas)")
ventas_productos_tienda4.plot(kind='bar', title='Ventas por producto - Tienda 4')
plt.xlabel('Producto')
plt.ylabel('Cantidad de ventas')
plt.show()
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda

envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f'Envío promedio Tienda 1: ${envio_promedio_tienda1:.2f}')
print(f'Envío promedio Tienda 2: ${envio_promedio_tienda2:.2f}')
print(f'Envío promedio Tienda 3: ${envio_promedio_tienda3:.2f}')
print(f'Envío promedio Tienda 4: ${envio_promedio_tienda4:.2f}')
```

  - **Tienda 4: $23,459.46 (el más bajo)**  

- **Ventas por categoría**:  
  Todas las tiendas tienen mayor demanda en **Muebles y Electrónicos**, pero la Tienda 4 muestra menor dinamismo en categorías secundarias.  

- **Productos más y menos vendidos en Tienda 4**:  
  - Más vendido: *Cama box* (62 ventas)  
  - Menos vendido: *Guitarra eléctrica* (33 ventas)  

---

## Conclusión
Considerando todos los factores, la **Tienda 4 es la candidata más adecuada para ser vendida**.  

- Es la de **menores ingresos totales**,  
- Su ventaja en **costos de envío** no se traduce en un mayor volumen de ventas,  
- Y aunque su calificación es aceptable (4.00), no sobresale frente a las demás.  

### Recomendación
El Sr. Juan debería **poner en venta la Tienda 4**, destinando los recursos obtenidos a invertir en un nuevo emprendimiento con mayor potencial de crecimiento.
